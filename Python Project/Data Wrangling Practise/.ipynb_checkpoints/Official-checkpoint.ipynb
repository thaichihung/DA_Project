```python
# import library
import pandas as pd
import numpy as np
import os
```

```python
# import file
dataframe_prices = pd.read_csv('prices.csv')
dataframe_sales = pd.read_csv('sales.csv')
```

```python
# convert object to date and sort 2 dataframe and check for null value
dataframe_prices['updated_at'] = pd.to_datetime(dataframe_prices['updated_at'])
dataframe_sales['ordered_at'] = pd.to_datetime(dataframe_sales['ordered_at'])
dataframe_prices
```

```python
# join 2 tables and find the revenue for each product_id at each ordered time
distinct_product_id = list(set(dataframe_prices['product_id']))
frames = []
for i in distinct_product_id:
    df_sales=dataframe_sales[dataframe_sales['product_id'] == i]
    df_prices=dataframe_prices[dataframe_prices['product_id'] == i]
    prices_sales = pd.merge_asof(df_sales.sort_values(['ordered_at']), df_prices.sort_values(['updated_at']), 
                                     left_on = 'ordered_at', right_on = 'updated_at', suffixes = ('_sales','_prices'), 
                                             direction = 'nearest')
    frames.append(prices_sales)
result = pd.concat(frames)
result['revenue']=np.where(result['ordered_at']>result['updated_at'], result['new_price'], result['old_price'])
result
```

```python
#calculate total quantirty_ordered and total_revenue
final_df = result[['product_id_sales','quantity_ordered','revenue']]
final_df = final_df.groupby(['product_id_sales','revenue'], as_index=False)['quantity_ordered'].sum()
final_df['total_revenue'] = final_df['revenue']*final_df['quantity_ordered']
final_df
```
